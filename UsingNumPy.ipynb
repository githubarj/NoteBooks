# Indexing and slicing of arrays

```python
e = np.array([
        #index  0   1  2  3
                [1, 2, 3, 4], # index 0
                [3, 4, 5, 6], # index 1
                [6, 7, 8, 9]  # index 2
])
```

```python
e.shape
```

```python
# to get the entire row of an array
e[0]
```

```python
# to get a pecific element you first specify the index of the row in which it is and then the column
e[0][1]
```

```python
# a more efficient way to do it is with one bracket
e[0, 1] 
```

# Slicing Matrices

```python
# we can also select a groupof things in a matrice. like the first two entire rows. 
e[0:2] # This will select the index 0 row and index 1 row but not the row at index 2, this is because the second value i sthe uppper limit
```

```python
# This is to select everything in the matrix
e[:]
```

```python
# we can also select from different dimensions, for example to select all the rows but only two columns
e[:,0:2]
```

# Manipuating Arrays

```python
e[:]
```

```python
# if you are adding the same value, you can add as follows
e[2] = 15
# Lets check the array now
e[:]
```
