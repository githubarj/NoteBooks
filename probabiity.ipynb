```python
import matplotlib.pyplot as plt
from numpy.random import binomial,poisson,normal
import numpy as np
from scipy import stats
```

```python
n=7
p=0.45
x=np.array([*range(0,n+1)])
x=x.reshape(n+1,1)
```

```python
x.shape
```

```python
y=stats.binom.pmf(x,n,p)
```

```python
plt.plot(x,y,label=f"Binomial({n},{p})")
plt.xlabel("R.V X = # of girl child given birth to")
plt.ylabel("p(x)")
plt.legend()
```

```python
a,b=3,2
(stats.binom.cdf(a,n,p))-(stats.binom.cdf(b-1,n,p))
```

```python
a,b=1,0
(stats.binom.cdf(a,n,p))-(stats.binom.cdf(b-1,n,p))
```

```python
n=1000
p=0.49
x=np.array([*range(0,n+1)])
x=x.reshape(n+1,1)
```

```python
y=stats.binom.pmf(x,n,p)
```

```python
z=stats.poisson.pmf(x,n*p)
```

```python
j=stats.norm.pdf(x,n*p,n*p*(1-p))
```

```python
plt.plot(np.array(x),y,label=f'binomial({n},{p})')
plt.plot(x,z,label=f'poisson({n}{p})')
plt.plot(x,j,label=f'normal({n*p},{n*p*(1-p)})')
plt.legend()
```
